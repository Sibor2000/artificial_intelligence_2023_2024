<h4> comparing built in methods with a simple for loop function

```python
t = np.array(range(1, 1001))
```

```python
np.cumsum(t)
```

```python
%timeit np.cumsum(t)
```

```python
t = list(range(1,1001))
```

```python
def my_cumsum(my_vector):
    constructed = []
    total = 0
    for i in my_vector:
        total+=i
        constructed.append(total)
    return constructed
```

```python
my_cumsum(t)
```

```python
%timeit my_cumsum(t)
```
